Extracting

```python
## CSV Local Sources

db_09 = spark.read.csv('PNAD_COVID_092020.csv', sep = ',', inferSchema = True, header = True)
db_10 = spark.read.csv('PNAD_COVID_102020.csv', sep = ',', inferSchema = True, header = True)
db_11 = spark.read.csv('PNAD_COVID_112020.csv', sep = ',', inferSchema = True, header = True)

print('db_09.column count:', len(db_09.columns))
print('db_10.column count:', len(db_10.columns))
print('db_11.column count:', len(db_11.columns))

print('db_09.columns:', db_09.columns)
print('db_10.columns:', db_10.columns)
print('db_11.columns:', db_11.columns)
```

Transforming

```python
## Matching columns to append

db_11 = db_11.drop(*['A006A','A006B','A007A'])

print('db_09.count:', db_09.count())
print('db_10.count:', db_10.count())
print('db_11.count:', db_11.count())

print('db_09.column count:', len(db_09.columns))
print('db_10.column count:', len(db_10.columns))
print('db_11.column count:', len(db_11.columns))

print('db_09.columns:', db_09.columns)
print('db_10.columns:', db_10.columns)
print('db_11.columns:', db_11.columns)

print('db_09.dtypes:', db_09.dtypes)
print('db_10.dtypes:', db_10.dtypes)
print('db_11.dtypes:', db_11.dtypes)

print('db_09.schema:', db_09.schema)
print('db_10.schema:', db_10.schema)
print('db_11.schema:', db_11.schema)
```

```python
## Checking errors to append

db_09 = db_09.dropDuplicates()
db_10 = db_10.dropDuplicates()
db_11 = db_11.dropDuplicates()

# print('db_09.count:', db_09.count())
# print('db_10.count:', db_10.count())
# print('db_11.count:', db_11.count())

db_09.describe().show()
db_10.describe().show()
db_11.describe().show()

# db_09 = db_09.show(5)
# db_10 = db_10.show(5)
# db_11 = db_11.show(5)

# db_11.printSchema()
```

```python
## Appending dbs into one db

db_pnad = db_09.union(db_10)
db_pnad = db_pnad.union(db_11)
# db_pnad = db_09.append(db_10, ignore_index=True)

db_pnad.show(2)
db_pnad.printSchema()
print('db_pnad.count: ', db_pnad.count())
print('db_pnad column count: ', len(db_pnad.columns))
```

```python
# Filter columns
df = db_pnad.select(
    F.col("Ano").alias("ano"),
    F.col("UF").alias("uf"),
    F.col("CAPITAL").alias("capital"),
    F.col("RM_RIDE").alias("rm_ride"),
    F.col("V1008").alias("n_domicilio"),
    F.col("V1012").alias("n_semana"),
    F.col("V1013").alias("mes_pesquisa"),
    F.col("V1016").alias("n_entrevista_domicilio"),
    F.col("Estrato").alias("estrato"),
    F.col("UPA").alias("unidade_primaria_amostragem"),
    F.col("V1022").alias("situacao_domicilio"),
    F.col("V1023").alias("tipo_area"),
    F.col("V1030").alias("projecao_populacao"),
    F.col("V1031").alias("peso_domicilio_pessoas_com_pos"),
    F.col("V1032").alias("peso_domicilio_pessoas_sem_pos"),
    F.col("posest").alias("dominios_projecao"),

    # Parte A - Características gerais dos moradores
    F.col("A002").alias("idade_morador"),  # 1
    F.col("A003").alias("sexo"),           # 2

    # Parte B - COVID19 - Todos os moradores
    F.col("B0031").alias("ficou_em_casa"),                     # 3
    F.col("B0041").alias("buscou_atendimento_posto_ubs_esf"),  # 4
    F.col("B0042").alias("buscou_atendimento_ps_sus_upa"),     # 5
    F.col("B0043").alias("buscou_atendimento_hospital_sus"),   # 6
    F.col("B005").alias("ficou_internado"),                    # 7
    F.col("B006").alias("internado_risco"),                    # 8
    F.col("B007").alias("possui_plano_saude"),                 # 9

    # Parte C - Características de trabalho das pessoas de 14 anos ou mais de idade
    F.col("C011A").alias("respondeu_valor_recebido"),          # 10
    F.col("C011A1").alias("valores_recebidos_unidade"),        # 10
    F.col("C011A11").alias("valores_recebidos_faixa"),         # 10
    F.col("C011A12").alias("valores_recebidos_reais"),         # 10

)

df.show()
```
